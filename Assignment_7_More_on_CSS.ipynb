Q1) Create a navigation bar similar to the below-mentioned image. The navigation bar must contain 5 navigation links that are properly placed within the navigation bar using the CSS box model. The navigation items must change their text color on hover

```python
from IPython.display import display, HTML

html_code = """

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>navigation-bar</title>
    <style>
        *{
            margin: 0;
            padding: 0;
        }

        nav{
            padding:0px 75px;
            background-color: lightskyblue;
            height:100px;
        }
        ul {
            display:flex;
            padding:50px 0px 0px 0px;

        }
        ul li{
            margin:0px 25px;
        }
        li a{
            color:#fff;
        }
        li a:hover{
            color:red;
            transition:all 0.4s;
        }
    </style>
</head>
<body>
    <section class="section1">
        <div class="container">
            <nav>
                <ul type="none">
                    <li> <a target="_self" href="#">Home</a></li>
                    <li> <a target="_blank" href="#">About Us</a> </li>
                    <li><a target="_blank" href="#">Contact</a> </li>
                    <li><a target="_blank" href="#">Blog</a></li>
                    <li><a target="_blank" href="#">Login</a></li>
                </ul>
            </nav>
        </div>
    </section>

</body>
</html>
"""

# Display the HTML content
display(HTML(html_code))
```
